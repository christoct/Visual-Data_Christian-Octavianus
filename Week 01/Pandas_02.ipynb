# Persiapan Data Frame

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')
df= pd.DataFrame(np.random.randint(1,5,size=(n_rows,n_cols)),columns=cols)
df
```

# Selection dengan operator logika |

```python
df[(df['A']==1) | (df ['A']==3)]
```

# Selection dengan fungsi isin()

```python
df[df['A'].isin([1,3])]
```

# Mengenal operator negasi ~

```python
df[~df['A'].isin([1,3])]
```
